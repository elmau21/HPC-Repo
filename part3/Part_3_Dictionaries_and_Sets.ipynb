```python
mid_2010_set = set(mid_2010['CVEGEO'])
mid_2020_set = set(mid_2020['CVEGEO'])

remains_2010_2020 = mid_2010_set & mid_2010_set

new_in_2020 = mid_2020_set - mid_2010_set

disappears_2010_2020 = mid_2010_set - mid_2020_set
```

```python
print("mid_2010_set:", mid_2010_set)
print("mid_2020_set:", mid_2020_set)
```

```python
remains_2010_2020_gdf = mid_2020[mid_2020["CVEGEO"].isin(remains_2010_2020)]

new_in_2020_gdf = mid_2020[mid_2020["CVEGEO"].isin(new_in_2020)]

disappears_2010_2020_gdf = mid_2010[mid_2010["CVEGEO"].isin(disappears_2010_2020)]
```

```python
import matplotlib.pyplot as plt

import matplotlib.pyplot as plt

fig, ax = plt.subplots(1, 1, figsize=(10, 10))

# Plot the base layer
remains_2010_2020_gdf.plot(ax=ax, color='green', label='Remains 2010-2020')

# Plot new AGEBs on top
new_in_2020_gdf.plot(ax=ax, color='blue', label='New in 2020')

# Plot disappeared AGEBs
disappears_2010_2020_gdf.plot(ax=ax, color='red', label='Disappeared 2010-2020')

# Customizations
ax.set_title('AGEB Changes from 2010 to 2020')
plt.show()
```
